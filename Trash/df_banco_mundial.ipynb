```python
#Importamos las librerías que vamos a utilizar
#Banco Mundial con wbdata (PBI per cápita, gasto en educación, población)

import wbdata
import pandas as pd
import datetime
```

| Indicador                  | Código              |
| -------------------------- | ------------------- |
| PBI per cápita             | `NY.GDP.PCAP.CD`    |
| Gasto en educación (% PBI) | `SE.XPD.TOTL.GD.ZS` |
| Población total            | `SP.POP.TOTL`       |


```python
#Crear diccionario de indicadores

indicadores = {
    'NY.GDP.PCAP.CD': 'pbi_per_capita',
    'SE.XPD.TOTL.GD.ZS': 'gasto_educacion',
    'SP.POP.TOTL': 'poblacion'
}
```

```python
#Definir rango de años

fecha_inicio = datetime.datetime(2020, 1, 1)
fecha_fin = datetime.datetime(2024, 1, 1)
```

```python
#Descargar datos

df_wb = wbdata.get_dataframe(
    indicadores,
    date=(fecha_inicio, fecha_fin)
)
```

```python
#Convertir índices en columnas

df_wb = df_wb.reset_index()
```
